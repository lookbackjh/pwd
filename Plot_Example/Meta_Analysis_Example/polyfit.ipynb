## N=20 일때의 Sigma VS Bandwidth plotting


```python
x=read.table('new_sigma20.txt', header = FALSE,  sep = '	',  stringsAsFactors = FALSE)
y=read.table('new_bw20.txt',header=FALSE)
```

```python
df=data.frame(x,y)
colnames(df)=c("sigma","bw")
df$sigma2 <- df$sigma^2
quadraticModel <- lm(bw ~ sigma + sigma2, data=df)
summary(quadraticModel)
```

```python
sigma<- seq(0, 10, 0.1)
bw <- predict(quadraticModel,list(sigma=sigma, sigma2=sigma^2))
print(length(bw))
print(length(sigma))
```

```python
plot(df$sigma, df$bw, pch=16)
lines(sigma, bw, col='blue')
```

```python
res <- resid(quadraticModel)
qqnorm(res)
qqline(res)
```

```python
plot(fitted(quadraticModel), res)
abline(0,0)
```

```python
x=read.table('new_sigma_changed.txt', header = FALSE,  sep = '	',  stringsAsFactors = FALSE)
y=read.table('new_bw_changed.txt',header=FALSE)
dftot=data.frame(x,y)
colnames(dftot)=c("sigma","bw")
dftot$sigma2 <- dftot$sigma^2
quadraticModeltot <- lm(bw ~ sigma + sigma2, data=dftot)
summary(quadraticModeltot)
sigma<- seq(0, 10, 0.1)
bw <- predict(quadraticModeltot,list(sigma=sigma, sigma2=sigma^2))
```

```python
ggplot(dftot,aes(x=sigma, y=bw))+labs(size=4, theme(title=element_text(size=4,hjust=0.5)))+geom_point(colour="black", shape=1, size=2)+ylab(TeX(r'($log$(bandwidth))'))+xlab(TeX(r'($\sigma)'))+theme(axis.title.x = element_text(size = 18),axis.text.x=element_text(size=18))+theme(axis.title.y = element_text(size = 18),axis.text.y=element_text(size=18))+
      stat_smooth(aes(y = bw),method = "lm", formula = y ~ x + I(x^2), size = 1)
ggsave(filename="n600_k100_revised.png",width=3000,height=3000,unit="px")

dev.off()
```
