### Illustration of plotting the ordered point-wise entropy. 


```python
from cProfile import label
import math
import numpy as np
import matplotlib.pylab as plt
from collections import OrderedDict
def getentropy(prob):
    i=0
    feat_entropy=[]
    for arr in prob:
        cur=[]
        for temp in arr:
            if(temp!=0):
                cur.append(-math.log(temp))
            else:
                cur.append(99999)
        feat_entropy.append(cur)
    return feat_entropy    
def getaibindex(probdata):
    logprob=[]
    for arr in probdata:
        t_sum=0
        for ele in arr:
            if ele==0:
                continue
            else:
                t_sum+=(ele)
        logprob.append(t_sum/20)
    return logprob
def grouplot(s1,s3,k=-1,n=0,groupnum=0):
    vacteriainfo=np.arange(len(s1))
    mydict1=dict(zip(vacteriainfo,s1))
    mydict3=dict(zip(vacteriainfo,s3))
    colormap=['blue','green']
    myList1 = sorted(mydict3.items(), key=lambda item: item[1],reverse=True)
    x1, y1 = zip(*myList1[:])
    print(len(x1))
    t1=np.arange(len(x1))
    plt.figure(figsize=(10,10))
    plt.title('Ordered Pointwise Entropy for when n={}, k={} for group {}'.format(n,k,groupnum+1))
    plt.xticks(rotation='vertical')
    plt.xticks(fontsize=6)
    plt.ylim(0,15)
    secondinfo=[]
    thirdinfoi=[]
    for i in x1:
        secondinfo.append(mydict1[i])
    if groupnum>0:
        plt.plot(t1,secondinfo,color=colormap[1])
    else:
        plt.plot(t1,secondinfo,color=colormap[0])
    #plt.plot(t1, y1,label="avg",color='red')
    plt.legend()
    plt.show()
def avgplot(s1,s3,k=-1,n=0,groupnum=0):
    vacteriainfo=np.arange(len(s1))
    mydict1=dict(zip(vacteriainfo,s1))
    mydict3=dict(zip(vacteriainfo,s3))
    colormap=['red','blue','green']
    myList1 = sorted(mydict3.items(), key=lambda item: item[1],reverse=True)
    x1, y1 = zip(*myList1[:])
    print(len(x1))
    t1=np.arange(len(x1))
    plt.figure(figsize=(10,10))
    plt.title('Average of Ordered Pointwise Entropy for the two group generated when n={}, k={}'.format(n,k))
    plt.xticks(rotation='vertical')
    plt.xticks(fontsize=6)
    plt.ylim(0,15)
    secondinfo=[]
    thirdinfoi=[]
    for i in x1:
        secondinfo.append(mydict1[i])
    #plt.plot(t1,secondinfo,label="group1")
    plt.plot(t1, y1,color='red')
    plt.legend()
    plt.show()
```

k=0.1 refers that 10 percent of the samples are coming from same distribution.

```python
import numpy as np

def get_groupk(n,ratio):
    from scipy.stats import truncnorm
    bounds=[0,20]
    predecsion_mean=np.random.randint(1,13,n)
    predecision_sd=np.random.randint(1,6,n)
    pointwiseentropyg1=[]
    pointwiseentropyg2=[]
    np.random.normal(3,1,20)
    predecsion_mean=np.random.randint(1,13,int(ratio*n))
    predecision_sd=np.random.randint(1,6,int(ratio*n))
    diff_mean1=np.random.randint(1,13,int((1-ratio)*n))
    diff_mean2=np.random.randint(1,13,int((1-ratio)*n))
    diff_sig1=np.random.randint(1,6,int((1-ratio)*n))
    diff_sig2=np.random.randint(1,6,int((1-ratio)*n))
    for i in range(int(ratio*n)):
        loc=predecsion_mean[i]
        scale=predecision_sd[i]
        sample1=[]
        sample2=[]
        for sample in range(20):
            s1=truncnorm.rvs((bounds[0]-loc)/scale, (bounds[1]-loc)/scale, loc=loc, scale=scale)
            s2=truncnorm.rvs((bounds[0]-loc)/scale, (bounds[1]-loc)/scale, loc=loc, scale=scale)
            sample1.append(s1)
            sample2.append(s2)
        pointwiseentropyg1.append(sample1)
        pointwiseentropyg2.append(sample2)
    for i in range(int((1-ratio)*n)):
        loc1=diff_mean1[i]
        scale1=diff_sig1[i]
        loc2=diff_mean2[i]
        scale2=diff_sig2[i]
        sample1=[]
        sample2=[]
        for sample in range(20):
            s1=truncnorm.rvs((bounds[0]-loc)/scale, (bounds[1]-loc)/scale, loc=loc1, scale=scale1)
            s2=truncnorm.rvs((bounds[0]-loc)/scale, (bounds[1]-loc)/scale, loc=loc2, scale=scale2)
            sample1.append(s1)
            sample2.append(s2)
        pointwiseentropyg1.append(sample1)
        pointwiseentropyg2.append(sample2)

    p1_10=np.array(pointwiseentropyg1)
    p2_10=np.array(pointwiseentropyg2)
    return p1_10,p2_10
p1_10,p2_10=get_groupk(600,0.1)
p1_50,p2_50=get_groupk(600,0.5)
p1_90,p2_90=get_groupk(600,0.9)
avglog1=getaibindex(p1_90)
avglog2=getaibindex(p2_90)
avglog3=[(g + h) / 2 for g, h in zip(avglog1, avglog2)]
avgplot(avglog1,avglog3,k=0.9,n=600,groupnum=0) ## here, k=0.9 means that 90 percent of the microbiomes are coming from same distribution. 
grouplot(avglog1,avglog3,k=0.9,n=600,groupnum=0)
grouplot(avglog2,avglog3,k=0.9,n=600,groupnum=1)
```
